### 4. Area Plot

Pada area plot ini akan ditampilkan 5 negara skala nasional dengan total indeks tertinggi

```python
#Memilih top 5 negara dengan level = nasional
data_top5 = sampleNational.sort_values(['Total'], ascending=False, axis=0)
condition = data_top5['Level'] == 'National'

data_top5 = data_top5[condition].tail()

#Menghapus kolom yang tidak diperlukan
data_top5.drop(['Level','Region', 'Total'], axis=1, inplace=True)

data_top5
```

```python
# Transpose data
data_top5 = data_top5.transpose()

data_top5.head()
```

Membuat area plot untuk 5 negara skala nasional dengan total indeks tertinggi

```python
# Mengubah tipe data nilai indeks dari data_top5 menjadi integer untuk plotting
data_top5.index = data_top5.index.map(int)

data_top5.plot(kind='area',
             stacked=False,
             figsize=(20, 10))  # pass a tuple (x, y) size

plt.title('Trend dari 5 negara dengan total indeks terendah')
plt.xlabel('Tahun')
plt.ylabel('Indeks')

plt.show()
```

Area plot di atas dapat kita atur transparansinya dengan menggunakan parameter `alpha`

```python
data_top5.plot(kind = 'area',
             stacked = False,
             alpha = 0.25,
             figsize = (20, 10))  # pass a tuple (x, y) size

plt.title('Trend dari 5 negara dengan total indeks terendah')
plt.xlabel('Tahun')
plt.ylabel('Indeks')

plt.show()
```
