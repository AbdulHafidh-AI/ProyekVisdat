### 2 FILTERING

```python
sampleNational = sample.set_index('Country')
sampleNational.head()
```

Menghapus Kolom yang tidak dibutuhkan

```python
sampleNational.drop(['GDLCODE','ISO_Code'], inplace=True, axis=1)
sampleNational.head()
```

Melihat Human Development Index pada tahun 1990 sampai 2019 di Indonesia

```python
sampleNational.loc['Indonesia', years]
```

```python
## Ambil yang Country = Indonesia dan membuat kondisi Level = National
National = sampleNational.loc['Indonesia']

condition = National['Level'] == 'National'
National = National[condition]

National
```

### 3. LINE PLOT

Menampilkan Line plot yang levelnya national

```python
result = National.loc['Indonesia', years]

result.plot()
```

```python
result.index = result.index.map(int)
result.plot(kind = 'line')

plt.title('Human Development Index in Indonesia')
plt.ylabel('Numbers of Index')
plt.xlabel('Year')
```

```python
df_result = sampleNational.loc[['Indonesia','Brazil']]
df_result
```

```python
condition = df_result['Level'] == 'National'
df_result = df_result[condition]
```

```python
df_result.plot(kind = 'line')
```

```python
df_result.head()
```

Menghapus kolom level dan region untuk plotting

```python
df_result.drop(['Level','Region','Total'], axis=1, inplace=True)

df_result.head()
```

```python
df_result = df_result.transpose()
df_result.head()
```

```python
df_result.index = df_result.index.map(int)
df_result.plot(kind='line')

plt.title('Human Development Index in Indonesia and Brazil')
plt.ylabel('Number of Index')
plt.xlabel('Years')

plt.show() 
```
